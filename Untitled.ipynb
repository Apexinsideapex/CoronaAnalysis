```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
```

```python
df = pd.read_csv('./train.csv')
df.head()
```

```python
df.drop(['Id','Province_State'], axis=1,inplace=True)
df['ConfirmedCases'] =df['ConfirmedCases'].astype(int)
df['Fatalities'] = df['Fatalities'].astype(int)
df.head()
```

```python
df.set_index(df['Country_Region'], inplace=True)
```

```python
df_total = df.groupby(by=['Country_Region']).max()
df_total.sort_values(by=['ConfirmedCases'],ascending=False,inplace=True)
df_total.head()
```

```python
top5_countries = list(df_total.iloc[:5].index)
```

```python
df_top5_cc = df.loc[top5_countries,['Date','ConfirmedCases']]
df_top5_ft = df.loc[top5_countries,['Date','Fatalities']]
df_top5_cc['Country'] = df_top5_cc.index
df_top5_cc.index = pd.RangeIndex(len(df_top5_cc))
df_top5_cc.head()
```

```python
df_top5_cc = pd.pivot_table(df_top5_cc, values=['ConfirmedCases'], index=['Date'], columns=['Country'],aggfunc='sum')
df_top5_cc = pd.DataFrame(df_top5_cc.to_records())
df_top5_cc.Date = df_top5_cc['Date'].astype('str')
df_top5_cc.set_index(df_top5_cc['Date'].astype('str'),inplace=True)
# df_top5_cc.drop(['Date'], axis=1, inplace=True)
```
